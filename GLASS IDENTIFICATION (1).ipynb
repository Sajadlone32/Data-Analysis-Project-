# 19.STORING THE FEATURES IN X AND TARGET VARIABLE IN Y

```python
x=df.drop('TYPE',axis=1)
y=df['TYPE']
```

# 20.SPLITING THE DATASET INTO TRAINING AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
df['TYPE'].value_counts()
```

# FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
```

```python
print('x_train.',x_train.shape)
print('y_train.',y_train.shape)
print('x_test.',x_test.shape)
print('y_test.',y_test.shape)
```

# 21.MODEL TRAINING

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

# Checking accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

# 22.Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred2=dtc.predict(x_test)
```

```python
#checking accuracy score
accuracy_score(y_test,y_pred2)
```

# 23. SVC

```python
from sklearn import svm
svc=svm.SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred3=svc.predict(x_test)
```

```python
#checking accuracy score
accuracy_score(y_test,y_pred3)
```

# 24.KNeighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
```

```python
knc.fit(x_train,y_train)
y_pred4=knc.predict(x_test)
```

```python
#checking accuracy score
accuracy_score(y_test,y_pred4)
```

# 25.Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
y_pred5=rfc.predict(x_test)
```

```python
#checking accuracy score
accuracy_score(y_test,y_pred5)
```

# 26.Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(x_train,y_train)
y_pred6=gbc.predict(x_test)
```

```python
 #checking accuracy score
accuracy_score(y_test,y_pred6)
```

# as we can see the above models the most accurate model is Random Forest Classier
