# 22. Seperating Features

```python
x=data.drop(columns=['Country','Happiness Score','Standard Error',],axis=1)
y=data['Happiness Score']
```

# 23.Creating Train test split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

# Model Training

# 24.logistic regression

```python
#training the logistic regression with training data
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred1=lr.predict(x_test)
```

# 25.Checking Accuracy of the model

```python
from sklearn.metrics import mean_squared_error, r2_score,explained_variance_score
```

```python
#checking accuracy score
mean_squared_error(y_test,y_pred1)
```

```python
#r2 score
r2_score(y_test,y_pred1)
```

```python
#explained_variance_score
explained_variance_score(y_test,y_pred1)
```

# 26.RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor()

rfr.fit(x_train,y_train)
y_pred2=rfr.predict(x_test)
```

```python
#cheking accuracy using mean r2 score
r2_score(y_test,y_pred2)
```

```python
#checking explained_variance_sore
explained_variance_score(y_test,y_pred2)
```

```python
#mean squard error
mean_squared_error(y_test,y_pred2)
```

# 27.Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr=GradientBoostingRegressor()

gbr.fit(x_train,y_train)
y_pred3=gbc.predict(x_test)
```

```python
#checking r2 score
r2_score(y_test,y_pred3)
```

```python
#cheking explained variance score
explained_variance_score(y_test,y_pred3)
```

```python
#checking mean squared error
mean_squared_error(y_test,y_pred3)
```

# 28.support vector regressor

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(x_train,y_train)
y_pred4=svr.predict(x_test)
```

```python
#checking r2 score
r2_score(y_test,y_pred4)
```

```python
#explained_variance_score
explained_variance_score(y_test,y_pred4)
```

```python
#checking mean squared error
mean_squared_error(y_test,y_pred4)
```
